```python
import pandas as pd
import plotly.express as px
import duckdb

# Cargar dataset
df = pd.read_csv("medical_insurance.csv")
```

```python
df.info()        # estructura del dataset
df.describe()    # estadísticas básicas
df.isnull().sum() # valores faltantes
```

```python
# Distribución de edades
px.histogram(df, x="age", nbins=20, title="Distribución de edades")

# Relación entre BMI y gastos
px.scatter(df, x="bmi", y="expenses", color="gender", title="BMI vs Gastos de seguro")

# Promedio de gastos por región
px.bar(df.groupby("region")["expenses"].mean().reset_index(),
       x="region", y="expenses", title="Gasto promedio por región")
```

```python
# Promedio de gastos por género
duckdb.query("SELECT gender, AVG(expenses) AS promedio_gasto FROM df GROUP BY gender")

# Promedio de gastos por región
duckdb.query("SELECT region, AVG(expenses) AS promedio_gasto FROM df GROUP BY region")

# Relación entre elegibilidad de descuento y gastos
duckdb.query("SELECT discount_eligibility, AVG(expenses) AS promedio_gasto FROM df GROUP BY discount_eligibility")
```

### Conclusión sobre descuento
- El promedio de gastos de quienes tienen descuento es 32,050.
- El promedio de gastos de quienes no tienen descuento es 8,434.
- Se observa que la elegibilidad de descuento influye en el gasto promedio.


```python
px.scatter(df, x="bmi", y="expenses", color="gender", title="BMI vs Gastos de seguro")
```

### Conclusión sobre BMI
- Se observa que a mayor BMI, los gastos de seguro tienden a ser más altos.

### Conclusiones finales

- **Género:** Se observa una diferencia en el gasto promedio entre hombres y mujeres. Según los resultados, [hombres/mujeres] presentan un gasto mayor.
- **Región:** La región con mayor gasto promedio es el sureste, lo que indica una variación geográfica importante en los costos de seguro.
- **BMI:** El gráfico muestra que a mayor índice de masa corporal (BMI), los gastos de seguro tienden a incrementarse, reflejando una relación positiva.
- **Descuento:** La elegibilidad de descuento influye claramente en el gasto promedio. Los asegurados con descuento presentan un gasto promedio distinto a los que no lo tienen.
